```python
# libraries 
from binance.client import Client
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Connect to binance API
INPUT_FILE= "API_KEYS.txt"

with open(INPUT_FILE, 'r', encoding='utf-8') as file:
    contenu_liste = file.read()

KEYS = contenu_liste.split(',')
KEYS = [element.strip() for element in KEYS]

api_key = KEYS[0]
api_secret = KEYS[1]
client = Client(api_key,api_secret)
```

```python
exchange_info = client.get_exchange_info()
# binance_pairs = [s['symbol'] for s in exchange_info['symbols'] if s['quoteAsset'] in ['USDT','USDC']]
binance_pairs = ['BTCUSDT', 'ETHUSDT', 'USDCUSDT', 'ATOMUSDT', 'DOGEUSDT', 'SOLUSDT','BTCUSDC']

# Initialisation du DataFrame pour stocker les données
df_trades = pd.DataFrame()

# Récupération des transactions pour chaque paire
for pair in binance_pairs:
    trades = client.get_my_trades(symbol=pair)  # Récupérer les trades pour la paire
    if trades:  # Vérifier s'il y a des trades
        pair_df = pd.DataFrame(trades)  # Convertir les trades de la paire en DataFrame
        pair_df['symbol'] = pair  # Ajouter la colonne du symbole pour chaque trade
        df_trades = pd.concat([df_trades, pair_df], ignore_index=True)  # Ajouter au DataFrame global

# Affichage des résultats
df_trades
```
