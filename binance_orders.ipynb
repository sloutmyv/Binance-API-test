```python
from binance.client import Client
from binance.exceptions import BinanceAPIException, BinanceRequestException
from datetime import datetime, timezone
import pandas as pd
```

```python
INPUT_FILE= "API_KEYS.txt"

with open(INPUT_FILE, 'r', encoding='utf-8') as file:
    contenu_liste = file.read()

KEYS = contenu_liste.split(',')
KEYS = [element.strip() for element in KEYS]

api_key = KEYS[0]
api_secret = KEYS[1]
client = Client(api_key,api_secret)
```

```python
# Current price of an Asset 
base_asset = 'BTC'
quote_asset = 'USDT'
symbol = base_asset + quote_asset

# balance_base_asset = client.get_asset_balance(asset=base_asset)
# balance_quote_asset = client.get_asset_balance(asset=quote_asset)
avg_price = client.get_avg_price(symbol=symbol)

print(f"{symbol} price is : {avg_price['price']} on {datetime.fromtimestamp((avg_price['closeTime']/1000), tz=timezone.utc).strftime('%Y-%m-%d %H:%M:%S')}")
```

```python
# Functions 
# Place a limit buy order (quantity : base currency and price : quote currency)
def place_limit_buy_order(client, symbol, quantity, price):
    """
    Place a limit buy order on Binance.

    Parameters:
        client: Binance client object.
        symbol: str, trading pair symbol (e.g., 'BTCUSDT').
        quantity: float, amount of the base asset to buy (e.g., 'BTC').
        price: str, price at which to buy the asset (e.g., 'USDT').

    Returns:
        dict: The response from Binance if the order is successful.
        None: If an exception occurs.
    """
    try:
        order = client.order_limit_buy(symbol=symbol, quantity=quantity, price=price)
        print("Order cancelled successfully.")
        return order
    except Exception as e:
        print(f"Failed to cancel the order: {e}")
        return None

def cancel_order(client, symbol, order_id):
    """
    Cancel an existing order on Binance.

    Parameters:
        client: Binance client object.
        symbol: str, trading pair symbol (e.g., 'BTCUSDT').
        order_id: str or int, the ID of the order to cancel.

    Returns:
        dict: The response from Binance if the cancellation is successful.
        None: If an exception occurs.
    """
    try:
        result = client.cancel_order(symbol=symbol, orderId=order_id)
        print("Order cancelled successfully.")
        return result
    except Exception as e:
        print(f"Failed to cancel the order: {e}")
        return None
```
